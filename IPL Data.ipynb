```python
#number of matches played at each venue
plt.figure(figsize=(12,6))
sns.countplot(x='venue' ,data=matches)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#number of matches played at each city
plt.figure(figsize=(12,6))
sns.countplot(x='city', data=matches)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#number of matches won by each team
plt.figure(figsize=(12,6))
sns.countplot(x='winner', data=matches)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#toss won by each team
plt.figure(figsize=(25,6))
sns.countplot(x='toss_winner', data=matches)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#number of matches played each ipl season - toss decision


count_series = matches.toss_decision.value_counts()
size = (count_series/count_series.sum())*100
plt.pie(size, labels = (np.array(count_series.index)), shadow = True)
```

```python
#number of matches played each ipl season - toss decision
plt.figure(figsize=(12,6))
sns.countplot(x='season',hue = 'toss_decision' ,data=matches)
plt.xticks(rotation='vertical')
plt.show()
```
